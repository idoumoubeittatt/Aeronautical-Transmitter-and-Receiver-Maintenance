```python
import pandas as pd
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test

# Créer un dataframe avec les données de survie
data = pd.DataFrame({
    'temps': [6, 7, 10, 12, 15, 20, 22, 35, 40, 50],
    'censure': [1, 0, 0, 1, 1, 0, 1, 0, 1, 1],
    'groupe': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B']
})

# Créer une instance du modèle de survie de Kaplan-Meier
kmf = KaplanMeierFitter()

# Diviser les données par groupe
group_A = data['groupe'] == 'A'
group_B = data['groupe'] == 'B'

# Adapter le modèle aux données du groupe A
kmf.fit(data['temps'][group_A], data['censure'][group_A], label='Groupe A')

# Tracer la courbe de survie du groupe A
kmf.plot()

# Adapter le modèle aux données du groupe B
kmf.fit(data['temps'][group_B], data['censure'][group_B], label='Groupe B')

# Tracer la courbe de survie du groupe B
kmf.plot()

# Effectuer un test de log-rank pour comparer les deux groupes
results = logrank_test(data['temps'][group_A], data['temps'][group_B],
                       event_observed_A=data['censure'][group_A],
                       event_observed_B=data['censure'][group_B])

# Afficher les résultats du test de log-rank
print(results.summary)
```
